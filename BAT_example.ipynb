## Poisson Counting Example

We want to determine the porperties of a radioactive signal source in the presence of background from natural sources of radioactivity. We assume to have one signal source $S$ and one background source $B$.

### 1. Background only measurement
- $k_\text{B} = 10$ counts with detector and without source
- Determine event rate of natural radioactive background

Define the log-likelihood using the function `logpdf()` and `Poisson()`from `Distributions.jl`.

```julia
# Number of observed background events
k_B = 10

likelihood_B = let k = k_B
    params -> begin
        return LogDVal(logpdf(Poisson(params.λ_B), k))
    end
end
```

Define the prior with the `NamedTupleDist()` function. We choose a flat prior between 0 and 30.

Afterwards, we use the likelihood and prior to define the `PosteriorDensity()`.

```julia
prior_B = BAT.NamedTupleDist(
    λ_B = Uniform(0, 30)
)

posterior_B = PosteriorDensity(likelihood_B, prior_B);
```

Define the settings for the sampling algorithm: We choose the `MetropolisHastings()` algorithm with $10^6$ samples and $4$ chains.

```julia
algorithm = MetropolisHastings()
nsamples = 10^5
nchains = 4
```

Calling `bat_sample()` to explore the parameter space of the posterior.

```julia
samples_B, chains_B = bat_sample(
    posterior_B,
    MCMCSampling(mcalg = algorithm, nsteps = nsamples, nchains = nchains)
);
```

Print some statistics of the samples.

```julia
sd_B = SampledDensity(posterior_B, samples_B)
display(sd_B)
```

Take a look at the resulting distribution of the background rate using `plot()`.

```julia
plot(samples_B, :λ_B)
plot!(prior_B, label = "Prior")
xlabel!("\$\\lambda_B\$")
ylabel!("\$p(\\lambda_B)\$")
```

### 2. Signal + Background

- $k_{\text{S+B}} = 12$ counts using the radioactive source in the setup
- Determine the signal rate $\lambda_{\text{S}}$ using the knowledge about the background

Define the likelihood for the signal + background model.

```julia
# Number of observed events
k_SB = 12

likelihood_SB = let k = k_SB
    params -> begin
        return LogDVal(logpdf(Poisson(params.λ_B + params.λ_S), k))
    end
end
```

Define the prior for both the signal and background parameters. The signal prior is chosen flat, while we take the background prior as distribution from the previous measurement.

```julia
hist = fit(Histogram, 
    BAT.flatview(samples_B.v.λ_B), 
    BAT.FrequencyWeights(BAT.flatview(samples_B.weight)), 
    nbins = 200
)

hist_B = UvBinnedDist(hist);
```

```julia
prior_SB = BAT.NamedTupleDist(
    λ_B = hist_B,
    λ_S = 0..30
)

posterior_SB = PosteriorDensity(likelihood_SB, prior_SB);
```

```julia
samples_SB, chains_SB = bat_sample(posterior_SB,
    MCMCSampling(mcalg = algorithm, nsteps = nsamples, nchains = nchains, trafo = NoDensityTransform())
);
```

```julia
plot(samples_SB, 
    param_labels=["\\lambda_B", "\\lambda_S"]
)
```

Again, some statistics.

```julia
println("Mode: $(mode(samples_SB))")
println("Mean: $(mean(samples_SB))")
println("Variance: $(var(samples_SB))")
println("Standard Deviation: λ_B = $(sqrt(var(samples_SB).λ_B[1])), λ_S = $(sqrt(var(samples_SB).λ_S[1]))")
```
